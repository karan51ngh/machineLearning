##2.Approach 1 : Label Encoding

###label encoding labels the different categories with whole numbers starting from **0** upto **number_of_classes - 1**

```python
from sklearn.preprocessing import LabelEncoder
```

```python
labelencoder_X = LabelEncoder()
```

```python
label_encoded_data = labelencoder_X.fit_transform(data['Remarks'])
```

```python
label_encoded_data
```

```python
label_encoded_data = pd.DataFrame(data = label_encoded_data, columns = ['Remarks'])
label_encoded_data
```
