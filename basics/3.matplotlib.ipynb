```
import numpy as np
import matplotlib.pyplot as plt
```

```
x=[2,5,5,6,7,8]
y=[3,7,7.2,4,12,5]

plt.plot(x,y)
plt.plot(x,y,'*')
```

```
# scatter
plt.scatter(x,y)
plt.xlabel('x axis')
```

```
# comparison
```

```
x1=[2,5,6,7,8]
y1=[3,7,4,12,5]

x2=[4,5,8,9,8]
y2=[3,5,7,12,6]

plt.plot(x1,y1,color='green',linewidth=3)
plt.plot(x2,y2,color='r')
plt.title('comparision')
plt.show()#hides stuff

# plt.annotate(xy=[4.5,4], s="YOLO")
```

```
x=np.arange(0,3*np.pi,0.1)
y=np.sin(x)
plt.plot(x,y)
```

```
# vertical bar graph

x1=[2,5,6,7,8]
y1=[3,7,4,12,5]

x2=[4,5,8,9,8]
y2=[3,5,7,12,6]

plt.bar(x1,y1,color='green',linewidth=3)
plt.bar(x2,y2,color='r',linewidth=3)
```
